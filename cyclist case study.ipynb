```python
# Merge all the 12 months data into one table
df=pd.concat([df5,df6,df7,df8,df9,df10,df11,df12,df1,df2,df3,df4])
df.head()
```

```python
# finding the null values
for col in df.columns:
    print(col,'-',np.sum(df[col].isnull()))
```

```python
# change the datatype of started_at to datetime64
df['started_at']=df['started_at'].astype('datetime64')
df['ended_at']=df['ended_at'].astype('datetime64')
```

```python
# find the ride duration in minutes
df['ride_duration']=(df['ended_at']-df['started_at'])/pd.Timedelta(minutes=1)
df['ride_duration']
```

```python
# finding the day of week
df['day']=df['started_at'].dt.weekday
```
